# Resizing images

```python
from torchvision import transforms
from torchvision.transforms import InterpolationMode as IM
from skimage.io import imread, imsave
import os
import torch
```

```python
transfs = transforms.Compose([transforms.Resize((224,224,), IM.BILINEAR)])
path = '../Data/train_images/'
save_path = '../Data/224/'
files = os.listdir(path)
```

```python
for F in files:
    pathFile = path + F
    img = transfs(torch.moveaxis(torch.Tensor(imread(pathFile)), -1, 0))
    imsave(save_path + F,torch.moveaxis(img, 0, -1))
```
